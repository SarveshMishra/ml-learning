## MatplotLib Anatomy Example

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2,3 ,4]
y = [112, 12, 34, 19]
```

```python
# 2. Setup plot
figure, axes = plt.subplots(figsize=(10, 10))
```

```python

# 3. Plot Data
axes.plot(x, y)
```

```python

# 4. Customize plot
axes.set(title="Simple Plot",
         xlabel="x-axis",
         ylabel="y-axis")
```

```python
# 5. Save
figure.savefig('../Step-3 Matplotlib/simple-plot.png')
```

## Making figures with NumPy array

We want :
- Line Plot
- Scatter Plot
- Bar Plot
- Histogram
- Subplot

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make plot form dictionary
nut_butter_prices = {
    "Almond" : 10,
    "Peanut" : 2,
    "Cashew": 3
}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="my price",
ylabel= "price",
xlabel="nuts")
```
